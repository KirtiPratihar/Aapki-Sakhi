```python
import csv
import pandas as pd
```

```python
def get_user_input():
    name = input("Enter your Full name: ")
    address = input("Enter your address: ")
    email = input("Enter your email address: ")
    phone_number = input("Enter your phone Name No: ")
    mailaddress = input("Enter your Mailing address: ")
    product_details = input("Enter your Product details: ")
    date_and_time = input("Enter the Preferred date and time: ")
    purpose_of_donation = input("Enter the Purpose of donation: ")
    
    
    return [name,address, email,phone_number,mailaddress,product_details,date_and_time,purpose_of_donation]

def write_to_csv(data, filename='collboration_database.csv'):
    with open(filename, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(data)

def main():
    filename = 'request_pad_data.csv'
    try:
        with open(filename, mode='x', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Name", "Address", "Email Id", "Phone Number","Mailing address","Product details","Preferred date and time"," Purpose of donation"])
    except FileExistsError:
        pass

    while True:
        user_data = get_user_input()
        write_to_csv(user_data)

        another = input("Do you want to enter another record? (yes/no): ").strip().lower()
        if another != 'yes':
            break

input_user = input("You want to add the data? ")
if input_user == 'yes': 
    if __name__ == "__main__":
        main()
elif input_user=="no":
    pass
else:
    print("Please inter the valid input")
```
